## 1. Viết chương trình biến đổi ảnh

### 1.1 Chọn đối tượng trong ảnh

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg')
bmg = data [800:1200, 570:980]
print (data.shape)
iio.imsave('orange.jpg', bmg)
plt.imshow (bmg)
plt.show ()
```

### Tịnh tiến đơn

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg', mode='L')
bdata = nd.shift (data, (100, 25))
plt.imshow (bdata)
plt.show ()
```

### 1.3 Thay đổi kích thước ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg')
print(data.shape)
bdata = nd.zoom(data, 2)
print(bdata.shape)
data2 = nd.zoom(data, (2, 2, 1))
print(data2.shape)
data3 = nd.zoom(data, (0.5, 0.9, 1))
plt.imshow(data3)
plt.show()
```

### 1.4 Xoay ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg')
print(data.shape)
d1 = nd.rotate(data, 20)
plt.imshow(d1)
plt.show()
d2 = nd.rotate(data, 20, reshape=False)
plt.imshow(d2)
plt.show()
```

### 1.5 Dilation and Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('world_cup.jpg', mode='L')
print (data.shape)
d1 = nd.binary_dilation (data)
plt.imshow (d1)
plt.show ()
d2 = nd.binary_dilation (data, iterations=3)
plt.imshow (d2)
plt.show()
e1 = nd.binary_erosion(data)
plt.imshow(e1)
plt.show ()

```

### 1.6 Coordinate Mapping

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('world_cup.jpg', mode='L')
print (data.shape)
V, H= data.shape
M = np.indices((V, H))
d = 5
q=2 * d * np.random.ranf (M.shape) - d
mp = (M + q).astype (int)
d1 = nd.map_coordinates (data, mp)
plt.imshow (d1)
plt.show ()
```

### 1.7 Biến đổi chung (Generic Transformation)

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0]/10.0 + outcoord[0])
    b = 10 * np.cos(outcoord[1]/10.0 + outcoord[1])
    return a, b
data = iio.imread('world_cup.jpg', mode = 'L')
d1 = nd.geometric_transform(data, GeoFun)
plt.imshow(d1)
plt.show()
```

### Bài 5

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt
images = {"1": "exercise/colorful-ripe-tropical-fruits.jpg", "2": "exercise/quang_ninh.jpg", "3": "exercise/pagoda.jpg"}
print("T - Tịnh tiến")
print("X - Xoay")
print("P - Phóng to")
print("H - Thu nhỏ")
print("C - Coordinate Map")
chon = input("Chọn thao tác (T/X/P/H/C): ").upper()
hinh = input("Chọn hình (1/2/3): ")
if hinh not in images:
    print("Không có ảnh!")
    exit()
data = iio.imread(images[hinh])
if chon == 'T':
    dx = int(input("Nhập chiều ngang: "))
    dy = int(input("Nhập chiều dọc: "))
    bdata = nd.shift(data, (dy, dx, 0))
    plt.imshow(bdata)
    plt.show()
elif chon == 'X':
    goc = float(input("Nhập góc xoay (độ): "))
    d1 = nd.rotate(data, goc)
    plt.imshow(d1)
    plt.show()
    d2 = nd.rotate(data, goc, reshape=False)
    plt.imshow(d2)
    plt.show()
elif chon == 'P':
    hs = float(input("Nhập hệ số phóng to: "))
    bdata = nd.zoom(data, (hs, hs, 1))
    plt.imshow(bdata)
    plt.show()
elif chon == 'H':
    hs = float(input("Nhập hệ số thu nhỏ: "))
    bdata = nd.zoom(data, (hs, hs, 1))
    plt.imshow(bdata)
    plt.show()
elif chon == 'C':
    d = float(input("Nhập độ biến dạng tọa độ: "))
    V, H, _ = data.shape
    M = np.indices((V, H))
    q = 2 * d * np.random.ranf((2, V, H)) - d
    mp = (M + q).astype(int)
    mp[0] = np.clip(mp[0], 0, V - 1)
    mp[1] = np.clip(mp[1], 0, H - 1)
    result = np.zeros_like(data)
    for i in range(3):
        result[:, :, i] = nd.map_coordinates(data[:, :, i], [mp[0], mp[1]])
    plt.imshow(result)
    plt.show()
else:
    print("Lựa chọn không hợp lệ.")
```
